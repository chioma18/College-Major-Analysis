```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
records = []
```

```python
for current_page in range(34):
    response = requests.get(f"https://www.payscale.com/college-salary-report/majors-that-pay-you-back/bachelors/page/{current_page + 1}")
    soup = BeautifulSoup(response.text, "html.parser")
    rows = soup.select("table data-table tbody tr")
    for row in rows:
      cells = row.select("span data-table__value")
      record = {
          "Undergraduate Major": cells[1].getText,
          "Starting Median Salary": float(cells[3].getText.strip("$")),
          "Mid-Career Median Salary": float(cells[4].getText.strip("$")),
      }
      records.append(record)
```

```python
pd.DataFrame(records).to_csv("salaries_by_college_major_updated.csv", index=False)
```
